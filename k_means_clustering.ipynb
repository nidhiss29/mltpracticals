```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
data=[[185,72],[170,56],[168,60],[179,68],[182,72],[188,77],[180,71],[180,70],[183,84],[180,88],[180,67],[177,76]]
df=pd.DataFrame(data,columns=["height","weight"])
df
```

```python
def eucledian(a,b):
  dis=math.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2)
  return dis
centroids=[[185,72],[170,56],[168,60]]
def compare(a,b):
  d=[]
  for i in b:
    z=eucledian(a,i)
    d.append(z)
  return (d.index(min(d))+1)
def k_means(a,c):
  x=a.iloc[:,:2].values
  r=[]
  for i in x:
    t=compare(i,c)
    r.append(t)
  return r
v=k_means(df,centroids)
print(v)
def visualize(df,k):
  x=df.iloc[:,0].values
  y=df.iloc[:,1].values
  for i in range(len(k)):
    if(k[i]==1):
      plt.scatter(x[i],y[i],color="red")
    elif(k[i]==2):
      plt.scatter(x[i],y[i],color="blue")
    elif(k[i]==3):
      plt.scatter(x[i],y[i],color="black")
  plt.show()
visualize(df,v)


  
  
```
