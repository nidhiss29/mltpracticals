```python
import pandas as pd
import numpy as np
def back(input,weights,bias,hidden,target):
  #for hidden layer
  q=[]
  hod=[]
  for i in range(0,len(weights),2):
    q.append(weights[i:i+2])
  for i in range(len(input)):
    z=0
    for k in range(len(q[1])):
      z=z+q[i][k]*input[k]
    h=z+bias
    h=1/(1+np.exp(-h))
    hod.append(np.round(h,2))
  hidden.append(hod)
  if(len(weights[4:])>0):
    back(np.round(hod,2),weights[4:],0.6,hidden,target)
    
  return hidden
def update(input,weights,bias,hidden,target,output):
    tra=output[0]*(1-output[0])
    tra3=output[1]*(1-output[1])
    #for hidden layer weights
    we=hidden[1]
    e=-(target[0]-output[0])
    e3=-(target[1]-output[1])
    w4,w6=[],[]
    for i in range(len(we)):
      op=1
      oi=1
      op*=we[0]*e*tra
      w4.append(op)
      oi*=we[1]*e3*tra3
      w6.append(oi)
    new=[]
    #for weight updation 
    weight_for_hidden=[]
    weight_for_hidden.append(weights[4:6])
    weight_for_hidden.append(weights[6:])
    for i in range(len(weight_for_hidden[1])):
      n=weight_for_hidden[0][i]-(w4[i]*0.5)
      new.append(n)
    for i in range(len(weight_for_hidden[1])):
      n=weight_for_hidden[1][i]-(w6[i]*0.5)
      new.append(n)
    #for input layer
    de=tra*e
    tra2=output[1]*(1-output[1])
    e2=-(target[1]-output[1])
    de2=tra2*e2
    rt=[]
    dr=de*(weight_for_hidden[0][0])
    rt.append(dr) 
    dr2=de2*(weight_for_hidden[1][0])
    rt.append(dr2)
    etotal=sum(rt)
    weig=[]
    weig.append(weights[:2])
    weig.append(weights[2:4])
    dhw1=input[0]
    dhw2=input[1]
    ewto=etotal*tra*dhw1
    ewt1=etotal*tra2*dhw2
    w1=[]
    for i in range(len(weight_for_hidden[1])):
      w=weig[0][i]-(ewto*0.5)
      w1.append(w)
    for i in range(2):
      w=weig[1][i]-(ewt1*0.5)
      w1.append(w)
    w1.extend(new)
    return w1
input=[0.05,0.10]
weights=[0.15,0.20,0.25,0.30,0.40,0.45,0.50,0.55]
target=[0.01,0.99]
c=0
a=back([0.05,0.10],[0.15,0.20,0.25,0.30,0.40,0.45,0.50,0.55],0.35,[[0.05,0.10]],[0.01,0.99])
while(a[-1]!=target):
  weights=update(input,weights,0.35,a,target,a[-1])
  a=(back(input,weights,0.35,[[0.05,0.10]],[0.01,0.99]))
  c=c+1
  print(a[-1])
print(c)
```
